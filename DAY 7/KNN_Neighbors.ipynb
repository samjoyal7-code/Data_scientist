```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
X = np.array([[2], [4], [6], [1]])
y = np.array(['Pass', 'Pass', 'Pass', 'Fail'])
```

```python
print(X)
print(y)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X,y)
```

```python
new_data = [[3]]
prediction = model.predict(new_data)
print("predicted result for 3 hours study:",prediction[0])
```

```python
plt.figure(figsize=(8,5))

for i in range (len(X)):
      if y [i] == 'Pass':
        plt .scatter(X[i],0,color='green',label='Pass'if i == 0 else"")
else:
        plt.scatter(X[i],0,color='red',label='Fail' if i == 3 else"")
plt.scatter(new_data,0,color='blue',marker='x',s=100,label='New Data(3 hours)')
plt.xlabel('Hours studied')
plt.yticks([])
plt.title('KNN - pass or Fail Prediction')
plt.legend()
plt.grid(True)
plt.show()
        
```
